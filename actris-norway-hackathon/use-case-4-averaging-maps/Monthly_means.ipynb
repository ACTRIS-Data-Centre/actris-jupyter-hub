```python
# Importing useful python libraries
import xarray as xr 
import threddsclient
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import requests
import json
```

```python
# Get the EBAS thredds catalog
all_opendap_urls = threddsclient.opendap_urls('https://thredds.nilu.no/thredds/catalog/ebas/catalog.html')
```

```python
# Get all filter_absorption_photometer opendap urls
fap_opendap_urls = [x for x in all_opendap_urls if 'filter_absorption_photometer' in x]
print('All EBAS filter absorption photometer datasets with opendap protocol: \n',fap_opendap_urls)
```

```python
opendap_urls = [x for x in fap_opendap_urls if 'NO0002' in x]
opendap_urls
```

```python
opendap_url = all_opendap_urls[10]
opendap_url
```

```python
ds = xr.open_dataset(opendap_url)
ds
```

```python
ds_monthly = ds.resample(time="M").mean()
ds_monthly
```
