```python
# Get the ACTRIS NRT thredds catalog
all_opendap_urls = threddsclient.opendap_urls('https://thredds.nilu.no/thredds/catalog/ebas/catalog.xml')
```

```python
pic_opendap_urls = [x for x in all_opendap_urls if 'NO0002' in x]
#print('All datasets from Zeppelin with opendap protocol: \n',pic_opendap_urls)

pic_opendap_urls = [x for x in pic_opendap_urls if 'picarro' in x]
#print('All picarro datasets from Zeppelin with opendap protocol: \n',pic_opendap_urls)

pic_opendap_urls = [x for x in pic_opendap_urls if '1.5' in x]
print('All NRT picarro datasets from Zeppelin with opendap protocol: \n',pic_opendap_urls)
```

```python
# For this example only one opendap_url is used: 
opendap_url = pic_opendap_urls[0]
opendap_url
```

```python
# Open and show dataset with xarray 
ds = xr.open_dataset(opendap_url)
ds
```

```python
ds_test = ds.sel(time=slice("2023-08-01", "2023-08-31"))
```

```python
ds_aug.carbon_monoxide_amean.plot()
```

```python
fig, ax = plt.subplots(2)
fig.set_figwidth(15)
fig.set_figheight(10)

for h in ds_aug.carbon_monoxide_amean.Tower_inlet_height.values:
    ds = ds_aug.carbon_monoxide_amean.sel(Tower_inlet_height=h)
    ax[0].plot(ds.time.values,ds.values, label = "Tower inlet heigh = {}".format(h))
    ax[0].set_xlabel('Time')
    ax[0].set_ylabel("{} [{}]".format(ds.standard_name.capitalize().replace("carbon_monoxide","CO").replace("_"," "), ds.units))
ax[0].legend()
ax[0].set_title("{} from {} at {}".format(
    ds_aug.carbon_monoxide_amean.ebas_component.capitalize().replace("_"," "),ds_aug.carbon_monoxide_amean.ebas_instrument_manufacturer, ds_aug.carbon_monoxide_amean.ebas_station_name))

for h in ds_aug.carbon_dioxide_amean.Tower_inlet_height.values:
    ds = ds_aug.carbon_dioxide_amean.sel(Tower_inlet_height=h)
    ax[1].plot(ds.time.values,ds.values, label = "Tower inlet heigh = {}".format(h))
    ax[1].set_xlabel('Time')
    ax[1].set_ylabel("{} [{}]".format(ds.standard_name.capitalize().replace("carbon_dioxide","CO2").replace("_"," "), ds.units))
ax[1].legend()
ax[1].set_title("{} from {} at {}".format(
    ds_aug.carbon_dioxide_amean.ebas_component.capitalize().replace("_"," "),ds_aug.carbon_dioxide_amean.ebas_instrument_manufacturer, ds_aug.carbon_dioxide_amean.ebas_station_name))



#ax[0].axvspan(ds_test.time.values[0],ds_test.time.values[-1], facecolor='orange', alpha=0.2, label = '29 Aug 2023')
#ax[1].axvspan(ds_test.time.values[0],ds_test.time.values[-1], facecolor='orange', alpha=0.2, label = '29 Aug 2023')

plt.tight_layout()
plt.legend()

plt.savefig('Birkenes_Picarro_CO_CO2_May.png')
plt.show()
```
