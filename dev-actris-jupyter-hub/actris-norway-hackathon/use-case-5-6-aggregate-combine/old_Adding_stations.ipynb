```python
# Importing useful python libraries
import xarray as xr 
import threddsclient
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import requests
import json
```

```python
# Get the EBAS thredds catalog
all_opendap_urls = threddsclient.opendap_urls('https://thredds.nilu.no/thredds/catalog/ebas/catalog.html')
```

```python
# Code to get all opendap urls from two stations in EBAS database.
opendap_urls = [x for x in all_opendap_urls if 'RO0007' in x or 'RO0010' in x]
print(opendap_urls)
```

```python
# Code to get all opendap urls from one station in EBAS database.
opendap_urls = [x for x in all_opendap_urls if 'NO0002' in x]
print(opendap_urls)
```

## Filter absorption photometer

```python
inst_opendap_urls = [x for x in opendap_urls if 'nephelometer' in x]
print('All Birkenes data from Nephelometers: \n',inst_opendap_urls)
```

```python
ds1 = xr.open_dataset(inst_opendap_urls[0])
ds2 = xr.open_dataset(inst_opendap_urls[1])
ds3 = xr.open_dataset(inst_opendap_urls[2])

ds1
```

```python
import xarray as xr

# Create an empty list to store the datasets
datasets = []

# Loop through each URL and open the dataset
for url in inst_opendap_urls:
    ds = xr.open_dataset(url)
    
    # Check if latitude and longitude match with other datasets
    for other_ds in datasets:
        if ds.geospatial_bounds != other_ds.geospatial_bounds:
            raise ValueError(f"Latitude and longitude do not match for dataset: {url}")
    
    # Append the dataset to the list
    datasets.append(ds)

# Merge all datasets into a new xarray dataset
merged_ds = xr.merge(datasets)
```
